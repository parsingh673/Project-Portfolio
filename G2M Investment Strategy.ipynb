**Hypothesis 1: 35-45 year olds in major cities around the United States create the most profit for cab companies. Yellow cab will appear to be the more profitable company.** 

```python
# Filtered the dataframe to include ages 35-45 and exclude incomes from 0-10000.
filtered_df = new_data[(new_data['Age'] >= 35) & (new_data['Age'] <= 45) & (new_data['Income (USD/Month)'] < 0) | (new_data['Income (USD/Month)'] > 10000)]

# Created a histogram of the income variable.
plt.hist(filtered_df['Income (USD/Month)'], bins = 10, color = 'orange')
plt.xlabel('Income (USD/Month)')
plt.ylabel('Frequency')
plt.title('Histogram of Income')
plt.show()

# Sought to see the distribution of income for the sample collected. To see different levels
# of income can give insight to where cab drivers should be in terms of location to pick up
# customers based off income of that area. 
```

```python
# Got the counts of each Customer ID.
customer_counts = filtered_df['Customer ID'].value_counts()

# Create a scatter plot of the Customer ID transaction counts.
plt.scatter(customer_counts.index, customer_counts.values, color = 'black')
plt.xlabel('Customer ID')
plt.ylabel('Number of Transactions')
plt.title('Scatter Plot of Customer Transactions')
plt.show()

# Able to see how many transactions each Customer ID had from the filtered dataframe.
# Allows me to see how often a customer would seek to ride in a taxi service and which
# Customer ID group specifically.
```

```python
# Defined the customer ID ranges.
customer_id_ranges = [range(0, 9001), range(17000, 22001), range(42000, 47001), range(52000, 56001), range(51000, 61001)]

# Combined the customer ID ranges into a single list.
customer_ids = [customer_id for customer_id_range in customer_id_ranges for customer_id in customer_id_range]

# Filtered the dataframe based on the selected customer IDs.
selected_data = new_data[new_data['Customer ID'].isin(customer_ids)]

# Selected the 'Price Charged' and 'City' columns.
selected_columns = ['Price Charged', 'Company', 'City']
selected_data = selected_data[selected_columns]

print(selected_data)

# Wanted to see specifically each Customer ID's 'Price Charged' and 'City' in order to target
# those high transaction counts from each Customer ID.
```

```python
# Grouped the data by 'Company' and calculated the sum of 'Price Charged'.
company_revenue = selected_data.groupby('Company')['Price Charged'].sum()

# Set the display format to print full numbers.
pd.options.display.float_format = '{:.2f}'.format

print(company_revenue)
```

**Reults of Hypothesis 1: Seeing from customer_data.describe the mean age of people using the taxi services was 35.36, the 50% quartile being 33 years old, and the 75% quartile being 42 years old I decided a sample from ages 35-45 would capture a significant portion of the data, including the middle 50% (from the 25th percentile to the 75th percentile) and a portion beyond that. Using the above analysis we saw how many transactions the specific groups had, then below that we created a scatter plot to see frequency of transactions for the filtered Customer IDs. We were able aggregate the Price Charged total per company based off high frequency use Customer IDs to see their revenues. Seeing that the Pink Cab company was so much lower in revenue this hypothesis leads to investing in the Yellow Cab company.**

```python
# Aggeregated 'Transaction ID' and then grouped by 'City'.
aggregated_tran_data = new_data.groupby('City')['Transaction ID'].count().reset_index()

# Ordered from highest to lowest based on 'Transaction ID'.
aggregated_tran_data = aggregated_tran_data.sort_values(by = 'Transaction ID', ascending = False)

print(aggregated_tran_data)
```

```python
# Merged City data into aggreagated transaction data frame to further confirm hypothesis. 
aggregated_trancity_data = pd.merge(aggregated_tran_data, city_data, on = 'City', how = 'left')

print(aggregated_trancity_data)
```

```python
# Grouped by 'City' and 'Company', counted 'Transaction ID', the sum of 'KM Travelled', and Profit by taking
# 'Price Charged' - 'Cost of Trip'.
aggregated_data = new_data.groupby(['City', 'Company']).agg({'Transaction ID': 'count', 'KM Travelled': 'sum', 'Price Charged': lambda x: (x - new_data['Cost of Trip']).sum()}).reset_index()

# Sorted values by 'Company' and 'Transaction ID'.
aggregated_data = aggregated_data.sort_values(by = ['Company', 'Transaction ID'], ascending = [False, False])

# Renamed to 'Profit'
aggregated_data = aggregated_data.rename(columns = {'Price Charged': 'Profit'})

# Changed format of 'Profit' printed.
aggregated_data['Profit'] = aggregated_data['Profit'].apply(lambda x: '{:,.2f}'.format(x))

print(aggregated_data)
```

It held true that high number of transactions leads to a higher number of cab users within the city but the disparity between Yellow Cab and Pink Cab companies was astonishing by analzing the Profit section, which took Price Charged - Cost of Trip. **Hypothesis 3: Based on a regression analysis using the variables 'KM Travelled' and 'Profit', it is anticipated that the Yellow Cab company will maintain higher profitability compared to the Pink Cab company.**

```python
new_data.head(10)
```

```python
# Made a new column 'Profit' by taking the same steps as before.
new_data['Profit'] = new_data['Price Charged'] - new_data['Cost of Trip']

# Got the sum of 'KM Travelled' for each Customer ID for the two cab companies. 
profit_km_data = new_data.groupby(['Customer ID', 'Company'])['KM Travelled'].sum().reset_index()

# Got the sum of 'Profit' generated from each Customer ID for the two cab companies.
profit_km_data['Profit'] = new_data.groupby(['Customer ID', 'Company'])['Profit'].sum().values
```

```python
profit_km_data.head(10)
```

```python
profit_km_data.head(-10)
```

```python
# Split the data into separate dataframes for each company.
yellow_cab_data = new_data[new_data['Company'] == 'Yellow Cab']
pink_cab_data = new_data[new_data['Company'] == 'Pink Cab']

# Created separate regression models for each company.
yellow_cab_model = LinearRegression()
pink_cab_model = LinearRegression()

# Fit the regression models using 'Profit' and 'KM Travelled' as predictors.
yellow_cab_model.fit(yellow_cab_data[['Profit', 'KM Travelled']], yellow_cab_data['Profit'])
pink_cab_model.fit(pink_cab_data[['Profit', 'KM Travelled']], pink_cab_data['Profit'])

# Predicted future profits for each company.
future_profit_yellow_cab = yellow_cab_model.predict(yellow_cab_data[['Profit', 'KM Travelled']])
future_profit_pink_cab = pink_cab_model.predict(pink_cab_data[['Profit', 'KM Travelled']])
```

```python
# Created a range of x values for plotting.
x = yellow_cab_data['KM Travelled']

# Plot the actual profits.
plt.scatter(x, yellow_cab_data['Profit'], color = 'orange', label = 'Actual Profits (Yellow Cab)')

# Plot the predicted future profits.
plt.plot(x, future_profit_yellow_cab, color = 'black', label = 'Predicted Future Profits (Yellow Cab)')

# Add labels and title to the plot.
plt.xlabel('KM Travelled')
plt.ylabel('Profit')
plt.title('Predicted Future Profits for Yellow Cab')
plt.legend()

plt.show()
```

```python
# Created a range of x values for plotting for Pink Company.
x = pink_cab_data['KM Travelled']

# Plot the actual profits.
plt.scatter(x, pink_cab_data['Profit'], color ='orange', label = 'Actual Profits (Pink Cab)')

# Plot the predicted future profits.
plt.plot(x, future_profit_pink_cab, color = 'black', label = 'Predicted Future Profits (Pink Cab)')

# Add labels and title to the plot.
plt.xlabel('KM Travelled')
plt.ylabel('Profit')
plt.title('Predicted Future Profits for Pink Cab')
plt.legend()

plt.show()
```

**Results of Hypothesis 3: As shown by the regression analysis, Yellow Cab will continue to make more profit on high KM riders than Pink Cab due to each Customer ID primarily having more of a higher aggregated KM Travelled for Yellow Cab.**

**Hypothesis 4: Regrssion Analysis between 'Cost of Trip' and 'KM Travelled' to visualize who will have more expenses in the future. I assume 'Cost of Trip' is a collective group of expenses a company has in order to find the right 'Price Charged' to charge to a Customer. I believe the Pink Cab Company will have less expenses than Yellow Cab Company.**

```python
# Aggregated 'Cost of Trip' expense account and 'KM Travelled' then grouped by 'Customer ID' and 'Company'.
company_expense = new_data.groupby(['Customer ID', 'Company']).agg({'Cost of Trip': 'sum', 'KM Travelled': 'sum'}).reset_index()

company_expense.head(10)
```

```python
# Split the data into separate dataframes for Yellow Cab specifically. 
yellow_cab_data2 = company_expense[company_expense['Company'] == 'Yellow Cab']

# Created a regression model for Yellow Cab.
yellow_cab_model2 = LinearRegression()

# Fit the regression model using 'Cost of Trip' and 'KM Travelled' as predictors.
yellow_cab_model2.fit(yellow_cab_data2[['Cost of Trip', 'KM Travelled']], yellow_cab_data2['Cost of Trip'])

# Predict future expenses for Yellow Cab.
future_expense_yellow_cab = yellow_cab_model2.predict(yellow_cab_data2[['Cost of Trip', 'KM Travelled']])

# Split the data into separate dataframes for Pink Cab.
pink_cab_data2 = company_expense[company_expense['Company'] == 'Pink Cab']

# Create a regression model for Pink Cab.
pink_cab_model2 = LinearRegression()

# Fit the regression model using 'Cost of Trip' and 'KM Travelled' as predictors.
pink_cab_model2.fit(pink_cab_data2[['Cost of Trip', 'KM Travelled']], pink_cab_data2['Cost of Trip'])

# Predict future expenses for Pink Cab.
future_expense_pink_cab = pink_cab_model2.predict(pink_cab_data2[['Cost of Trip', 'KM Travelled']])

# Determine the common axis limits in order to plot on the same scale.
x_min = min(yellow_cab_data2['KM Travelled'].min(), pink_cab_data2['KM Travelled'].min())
x_max = max(yellow_cab_data2['KM Travelled'].max(), pink_cab_data2['KM Travelled'].max())
y_min = min(future_expense_yellow_cab.min(), future_expense_pink_cab.min())
y_max = max(future_expense_yellow_cab.max(), future_expense_pink_cab.max())

# Plot both regression lines on the same scale values.
plt.figure(figsize = (10, 6))
sns.lineplot(x = yellow_cab_data2['KM Travelled'], y = future_expense_yellow_cab, label = 'Yellow Cab', color = 'orange')
sns.lineplot(x = pink_cab_data2['KM Travelled'], y = future_expense_pink_cab, label = 'Pink Cab', color = 'black')

plt.xlabel('KM Travelled')
plt.ylabel('Future Expense')
plt.title('Regression Analysis for Cab Expenses')
plt.legend()

# Set the common axis limits.
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

plt.show()
```

**To address the uneven distribution of data, I will refine my analysis by selecting a sample group and predicting the profits for each company within the excluded income range. My goal is to identify the middle-class segment where both companies can thrive in their business operations.**

```python
# Filtered the data based on the sample age range (35-45).
sample_analysis = new_data[(new_data['Age'] >= 35) & (new_data['Age'] <= 45)]

# Aggregated 'KM Travelled' and 'Profit' and grouped by 'Company' and 'Customer ID' columns.
sample_analysis = sample_analysis.groupby(['Customer ID', 'Company']).agg({'KM Travelled': 'sum', 'Profit': 'sum'}).reset_index()

# Print the result.
print(sample_analysis)
```

```python
# Wanted to see the disperse amongst the sample size by counting 'Customer ID' instead.
sample_count = sample_analysis.groupby('Company')['Customer ID'].count()

# Result shows a similar count compared to the company_count which confirms hypothesis's track.
print(sample_count)
```

```python
# Filter the data for Pink Cab.
pink_cab_data3 = sample_analysis[sample_analysis['Company'] == 'Pink Cab']

# Filter the data for Yellow Cab.
yellow_cab_data3 = sample_analysis[sample_analysis['Company'] == 'Yellow Cab']

# Create separate regression models for Pink Cab and Yellow Cab.
pink_cab_model3 = LinearRegression()
yellow_cab_model3 = LinearRegression()

# Fit the regression models using 'Profit' and 'KM Travelled' as predictors.
pink_cab_model3.fit(pink_cab_data3[['Profit', 'KM Travelled']], pink_cab_data3['Profit'])
yellow_cab_model3.fit(yellow_cab_data3[['Profit', 'KM Travelled']], yellow_cab_data3['Profit'])

# Predict the profits for Pink Cab and Yellow Cab.
pink_cab_predicted_profits = pink_cab_model3.predict(pink_cab_data3[['Profit', 'KM Travelled']])
yellow_cab_predicted_profits = yellow_cab_model3.predict(yellow_cab_data3[['Profit', 'KM Travelled']])

# Plot the regression lines for Pink Cab and Yellow Cab.
plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'KM Travelled', y = 'Profit', data = pink_cab_data3, alpha = 1, s = 50, label = 'Pink Cab')
sns.scatterplot(x = 'KM Travelled', y = 'Profit', data = yellow_cab_data3, alpha = 0.3, s = 50, label = 'Yellow Cab')
sns.lineplot(x = pink_cab_data3['KM Travelled'], y = pink_cab_predicted_profits, color = 'black')
sns.lineplot(x = yellow_cab_data3['KM Travelled'], y = yellow_cab_predicted_profits, color = 'orange')

# Plotted labels for visual.
plt.xlabel('KM Travelled')
plt.ylabel('Profit')
plt.title('Regression Analysis: Profit vs. KM Travelled')
plt.legend()
plt.show()
```

**Results of Hypothesis 5: Invest in Yellow Cab company. Through this analysis I provided fair parameters for both companies to adhere to and the result of the regression analysis shows that even if the sample may be equal in Customer IDs for both company, Yellow Cab company will continue to get more profits in the future.**
